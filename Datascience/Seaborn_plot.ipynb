```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

# Continuous Plot -
pairplot
jointplot
distplot

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
# it will draw pdf- probability density function, and draw best fit  regression line- bivariate analysis
```

```python
sns.pairplot(df)
# plotting for all combinaton of features- provided - all the features must be integer or floating - other features are ignored
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
df.loc[:,'smoker'].value_counts()
```

```python
# kind of histogram -density of tip - distributed plot
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
#kde is kernel density estimation
```

# Categorical plot
boxplot
violinplot
countplot
bar plot
shift+tab gives the attributes /description

```python
## countplot
sns.countplot('sex',data= df)
# Show the counts of observations in each categorical bin using bars
```

```python
sns.countplot(y='day',data=df)
```

```python
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
sns.barplot(x='smoker', y='total_bill', data=df)
```

```python
sns.barplot('sex', y='total_bill', data=df)
```

```python
sns.boxplot('sex','total_bill',data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker',data=df)
```

```python
sns.violinplot(x='total_bill', y='day',data=df)
```
