```python
# Importing Libraries

import numpy as np

import pandas as pd
```

```python
SpellBee_2022_df = pd.read_csv('Bee_Docs/Spell_Bee_Merged_db.csv')
```

```python
SpellBee_2022_df.sort_values(by='Words')
```

```python
dataset=SpellBee_2022_df.copy()
```

```python
pattern=['ceous','cious','seous','sious','ous','us','city','sity','tion','sion','cion','tive','cive','sive']
for match in pattern:
    dataset[match] = 0
    filt = dataset.loc[dataset.Words.str.endswith(match), :].index
    dataset.loc[filt,match] = 1
```

```python
pattern=['opp','app','aque','aqui','male','ph','x']
for match in pattern:
    dataset[match] = 0
    filt = dataset.loc[dataset.Words.str.startswith(match), :].index
    dataset.loc[filt,match] = 1
```

```python
dataset['WC_cious'] = 0
match='cious'
filt = dataset.loc[dataset.Words.str.endswith(match), :].index
dataset.loc[filt,'WC_cious'] = 1
dataset.loc[filt]
```
