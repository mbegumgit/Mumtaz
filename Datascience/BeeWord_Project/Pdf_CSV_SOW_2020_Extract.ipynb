```python
import tabula
import pandas as pd
```

# Extracting School of Words 2019- 2020  from decrypted pdf

```python
# While reading PDF specify the area correctly if the table is not in structured form . and also guess parameter for unstructured
Bee_2020_df_all= tabula.read_pdf("Bee_Docs/Bee_School Of Words_2020_decrypt.pdf",guess=False,pages='all',area=(8,5,96,100),relative_area=True)
```

To Update column names in each page extracted in a list . Data extracted from PDF contains list of dataframes each list 
mapped to each page in PDF

```python
#New_df = pd.DataFrame(columns=['Type','Description'])
#New_df[['Type','Description']] = Bee_2020_df_odd['Type'].str.split(')',expand=True)
def split_name(name): 
    split_name = str(name).split(')') 
    return split_name if len(split_name) == 2 else ['', split_name[0]]

def Append_pages(Bee_2020_df_all):
    Bee_Words_df=pd.DataFrame()
    for pg in range(len(Bee_2020_df_all)):
        if pg != 2:
            Bee_2020_df_all[pg].columns=['Index','Words','Type','Description']
            Bee_Words_df=pd.concat([Bee_Words_df,Bee_2020_df_all[pg]],axis=0,ignore_index=True)
        else:
            Bee_2020_df_all[pg].columns=['Index','Words','Type']
            Bee_Words_df=pd.concat([Bee_Words_df,Formatdf_odd(Bee_2020_df_all[pg])],axis=0,ignore_index=True) 
    return Bee_Words_df
             
def Formatdf_odd(Bee_2020_df_odd):
           
    New_df = pd.DataFrame(Bee_2020_df_odd['Type'].apply(split_name).tolist(), columns=['Type', 'Description'])
    New_df = pd.concat([Bee_2020_df_odd[['Index','Words']],New_df],axis=1,ignore_index=True)
    New_df.columns=['Index','Words','Type','Description']
    return New_df

Output_df = Append_pages(Bee_2020_df_all)
Output_df.dropna(how='all',inplace=True)
```

# First Parsing junk things from the Bee Word and then  upload to csv file

```python
Output_df['Words'].replace({'OR ':'OR xx'},inplace=True)
```

```python
import re
def Orsplit(x):
    match = re.findall('OR\s*\**([\w+\s*\w*]*)|\**(\w+((\s*|-)\w+)*)|\**([\w+\s*\w*]*)',x)
    for item in match[0]:
        if item != '':
            return item 
        
Output_df['Words'] = Output_df['Words'].dropna().apply(Orsplit)

Output_df.to_csv('Bee_Docs/2020_School of Words_Output.csv',encoding='utf-8-sig') # writing into csv format file
```
