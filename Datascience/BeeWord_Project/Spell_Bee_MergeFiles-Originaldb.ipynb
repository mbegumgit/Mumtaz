```python
SpellBee_2022_WoC_df = pd.read_csv('Bee_Docs/2022_Words Of Champions_Output.csv')
SpellBee_2022_School_df = pd.read_csv('Bee_Docs/2022_School of Words_Output_v1.csv')
```

Extracting only Words

```python
SpellBee_2022 =SpellBee_2022_WoC_df[['Difficulty Level','Words']]
SpellBee_2022['WOC_2022'] = 1
SpellBee_2022['SOW_2022'] = 0
```

```python
SpellBee_2022 = pd.merge(SpellBee_2022,SpellBee_2022_School_df['Words'],how="outer",on=['Words'])
```

```python
SpellBee_2022['Difficulty Level'].fillna('SOW',inplace=True)
filt = SpellBee_2022.loc[SpellBee_2022['Difficulty Level'] == 'SOW'].index
SpellBee_2022.loc[filt,'WOC_2022'] = 0
SpellBee_2022.loc[filt,'SOW_2022'] = 1
SpellBee_2022.loc[filt] 
```
