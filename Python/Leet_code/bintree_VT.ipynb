```python
# Vertical Order traversal of Binary Tree
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def __init__(self):
        self.point_list = []
        
    def dfs(self, root: TreeNode, x:int, y:int):
        self.point_list.append([x,y,root.val]) # keep in list all points
        print('point list:', self.point_list)
        print('what is the the pointer:',root.left, root.right)
        if root.left:
            print('Left')
            self.dfs(root.left, x-1, y-1)
            print('nodeL:', root.left)
        if root.right:
            self.dfs(root.right, x+1, y-1)
            print('nodeR:', root.right)
        
    def verticalTraversal(self, root: TreeNode) -> List[List[int]]:
        self.dfs(root, 0, 0)
        self.point_list.sort(key = lambda a:a[2]) # sort by value
        self.point_list.sort(key = lambda a:a[1], reverse = True) # sort by y
        self.point_list.sort(key = lambda a:a[0]) # sort by x
        print('Sorted:', self.point_list )
		
        res = []
        left = abs(self.point_list[0][0])
        print('L val', left, self.point_list[0], self.point_list[0][0])
        right = self.point_list[-1][0] 
        print('R val', right, self.point_list[-1]) 
        for i in range(left+right+1):
            res.append([j[2] for j in self.point_list if j[0]==i-left])
			
        return res 
            
        
```

```python
class Solution:
    def verticalTraversal(self, root: TreeNode) -> List[List[int]]:
        x_dict = defaultdict(list)
        
        # traverse the tree by level
        q = [(root, 0, 0)]
        while q:
            node, x, y = q.pop(0)
            print('each node:{} x:{}  y:{}'.format(node,x,y))
            x_dict[x].append((node.val, y))
            if node.left:
                q.append((node.left, x - 1, y - 1))
            if node.right:
                q.append((node.right, x + 1, y - 1))
        print(x_dict)
        ans = []
        # build the answer
        for x in sorted(x_dict.keys()):
            arr = []
            l, r = 0, 0
            while r < len(x_dict[x]) - 1:
                if x_dict[x][r+1][1] == x_dict[x][r][1]:
                    r += 1
                elif l == r:
                    arr.append(x_dict[x][l][0])
                    l += 1
                    r += 1
                else:
                    arr.extend([i[0] for i in sorted(x_dict[x][l:r+1])])
                    l = r + 1
                    r += 1
            if r >= l:
                arr.extend([i[0] for i in sorted(x_dict[x][l:])])
            
            ans.append(arr)
        return ans
```
